```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn import linear_model

```

```python
dataset=pd.read_csv("BostonHousePrices.csv")
```

```python
dataset.columns
```

```python
dataset
```

```python
dataset.describe()
```

```python
x=dataset['Rooms'] # independent variable
y=dataset['Value'] # dependent variable
```

```python
plt.figure(figsize=(12, 8), tight_layout=True)
sns.scatterplot(x=x, y=y)
plt.show()
```

```python
x = sm.add_constant(x) # adding a constant
lm = sm.OLS(y,x).fit() # fitting the model
```

```python
y_=lm.predict(x)
```

```python
lm.summary()
```

```python
# Rooms coef: 9.1021
# Constant coef: - 34.6706

# Linear equation: 𝑦 = 𝑎𝑥 + 𝑏
y_pred = 9.1021*x['Rooms'] - 34.6706
```

```python
# plotting the data points
plt.figure(figsize=(12, 8), tight_layout=True)
sns.scatterplot(x=x['Rooms'], y=y)

#plotting the line
sns.lineplot(x=x['Rooms'],y=y_pred, color='red')

#axes
plt.xlim(0)
plt.ylim(0)
plt.show()
```

#Multi-Linear Regression

```python
y = dataset['Value'] # dependent variable
X = dataset[['Rooms', 'Distance']] # independent variable
```

```python
lm = linear_model.LinearRegression()
lm.fit(X, y) # fitting the model
```

```python
y_pred = lm.predict(X)
```

```python
# plotting the data points
plt.figure(figsize=(12, 8), tight_layout=True)
sns.scatterplot(x=x['Rooms'], y=y)

#plotting the line
sns.lineplot(x=x['Rooms'],y=y_pred, color='red')

#axes
plt.xlim(0)
plt.ylim(0)
plt.show()
```

```python
# lm.predict(X)
a=0
input=a
price=lm.predict(X)[a:a+1]
print(X[a:a+1])
print('  predicted price = ', price)
print('  real price      = ', y[a:a+1])
```

```python
# r2 score
lm.score(X, y)
```

```python
lm.coef_
```

```python

lm.intercept_
```
